```python
import random
import numpy as np
from citipy import citipy
```

```python
#Generate a random integer
random.randint(-90,89)
```

```python
#Generate a random floating point value between 0 and 1
random.random()
```

```python
#Append the two random functions together to generate one number
random_number = random.randint(-90,89) + random.random()
random_number
```

```python
#Use uniform() function to generate a floating point decimal number 
#Group the resulting arrays as latitudes and longitudes
#Make number pairs from the two arrays
lats = np.random.uniform(-90, 90, size=1500)
long = np.random.uniform(-180, 180, size=1500)
lat_long = zip(lats, long)
lat_long
```

```python
#Add latitudes and longitudes to a list
cords = list(lat_long)
```

```python
#Use the print() function to display the latitude and longitude combinations
for coordinate in cords:
    print(coordinate[0], coordinate[1])
```

```python
# Use the print() function to display the nearest city
for coordinate in cords:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
#Create a list for holding cities
cities = []
#Identify the nearest city for each coordinate
for coordinate in cords:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    #If city is unique, add it to cities list
    if city not in cities:
        cities.append(city)
#Print the city count
len(cities)
```
